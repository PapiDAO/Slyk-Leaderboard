```python

import requests
from http.client import error
import urllib.request
import json
from urllib.request import  Request
from urllib.error import HTTPError



def process_id(id):
    """process a single ID"""
    # fetch the data
    r = requests.get("https://api.slyk.io/wallets/"+str(id)+"/balance?apikey=cES5DZL3B84Qxzt010eFmkDc0pSh1imI")
    # parse the JSON reply
    data = r.json()
    # and update some data with PUT

    return data

def process_range(id_range, store=None):
    """process a number of ids, storing the results in a dict"""
    if store is None:
        store = []
    for id in id_range:
         store.append({'id':id['id'], 'data': process_id(id['id'])['data'], 'ownerId':id['ownerId']})
    return store


from threading import Thread

def threaded_process_range(nthreads, id_range):
    """process the id range in a specified number of threads"""
    store = []
    threads = []
    # create the threads
    for i in range(nthreads):
        ids = id_range[i::nthreads]
        t = Thread(target=process_range, args=(ids,store))
        threads.append(t)

    # start the threads
    [ t.start() for t in threads ]
    # wait for the threads to finish
    [ t.join() for t in threads ]
    return store

def get_wallets(apikey, url="https://api.slyk.io/wallets?page[size]=100"):
    return_data = {}
    try:
            req = Request(url, headers={'User-Agent': 'Mozilla/5.0', 'apiKey': apikey})
            data = urllib.request.urlopen(req, timeout = 100)
            json_data = json.loads(data.read())
            return_data = json_data
            total_rows = json_data['total']
    except error as e:
            print(e)
    
    except HTTPError as e:
            print(e)
        
        
    for i in range(int(total_rows/100)):
        
        try:
            req = Request(url+"&page[number]="+str(i+2), headers={'User-Agent': 'Mozilla/5.0', 'apiKey': apikey})
            data = urllib.request.urlopen(req, timeout = 100)
            json_data = json.loads(data.read())
            return_data['data'].extend(json_data['data'])
            print(url+"&page[size]="+str(i+2))
            print(len(return_data['data']))
            
        except error as e:
            print(e)

        except HTTPError as e:
            print(e)
    return return_data


apikey="cES5DZL3B84Qxzt010eFmkDc0pSh1imI"
wallets =get_wallets(apikey)
wallets_ids = []
balance = []
for i in wallets['data']:
        wallets_ids.append({'id':str(i['id']), 'ownerId':i['ownerId']})
print(wallets_ids)

result = threaded_process_range(len(wallets_ids), wallets_ids)
print(result)
```
